# Inheritance

```python
class vehical:
    def fuel_type(self):
        print("fuel type is deisel")
class car(vehical):
    def max_speed(self):
        print("speed is 100km/h")
        
c=car()
c.max_speed()
c.fuel_type()
```

```python
class student:
    def __init__(self,name,branch,enroll):
        self.name=name
        self.branch=branch
        self.enroll=enroll
class personal(student):
    def __init__(self,n):
        self.n=n
        super().__init__("xyz","abc",123)
    def a(self):
        print("Hello")
    def display(self):
        print(self.name,self.branch,self.enroll)
        
p=personal(11)
print(p.name)
print(p.branch)
print(p.enroll)
p.display()
```

# Polymorphism

```python
# method overriding

class vehical:
    def fuel_type(self):
        print("fuel type is deisel")
    def max_speed(self):
        print("speed is 120km/h")
class car(vehical):
    def fuel_type(self):
        print("fuel type is patrol")
    def max_speed(self):
        print("speed is 100km/h")
        
c=car()
c.max_speed()
c.fuel_type()
```

```python
# method overloading
def adding(a,b,c):
    print(a+b+c)
def adding(a,b):
    print(a+b)
adding(1,2,3)
```

```python
# operator overloading
print("asd"+"fgh")
print(123+456)
print([1,2,3]+['a','b'])
```

```python
# operator overloading with custom object
class book:
    def __init__(self,p):
        self.p=p
    def __add__(self,*others):
        t=self.p
        for i in others:
            t+=i.p
        return book(t)
b1=book(100)
b2=book(200)
b3=book(500)
a=b1+b2+b3
print(a.p)
```

```python
class student:
    def __init__(self,m):
        self.m=m
    def __eq__(self,other):
        return self.m==other.m
    def __gt__(self,other):
        return self.m>other.m
s1=student(100)
s2=student(96)
if s1==s2:
    print("same")
else:
    print("not same")
if s1>s2:
    print("Greater than")
else:
    print("not greater")
```
