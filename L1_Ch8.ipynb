```python
class human:
    species ='Homosapiens'
    def __init__(self):
        self.name=input('Enter name: ')
        self.age=int(input('Enter age: '))
        self.gender=input('Enter gender: ')
    def speak(self):
        print("Hello, I can speak")
    def eat(self,favdish):
        self.f=favdish
        print('Hello, I love to eat ',self.f)
    def hello(self):
        print('I love ',self.f)
    def print_data(self):
        print("My name is ",self.name)
        print('My age is ',self.age)
        print('My gender is ',self.gender)
```

```python
h1= human()
h2= human()
print(h1.name)
print(h2.name)
print(h1.species)
h1.speak()
h1.eat('sweets')
h1.hello()
h1.print_data()
```
