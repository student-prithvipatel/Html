```python
from abc import ABC,abstractmethod
class Car(ABC):
    @abstractmethod
    def speed(self):
        print('faster')
class Tesla(Car):
    def speed(self):
        print('150 km/h')
class Maruti(Car):
    def speed(self):
        print('80 km/h')
t=Tesla()
t.speed()
```
