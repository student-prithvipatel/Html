```python
f=open('f1.txt','w') #w = write mode, writing, new file
f.write('Hello,Welcome \n')
l=['Jan\n','Feb\n','March']
f.writelines(l)
f.close()
```

```python
f1=open('f2.txt','a') #a= append mode, write in exiting file or make new file and write from start
f1.write('April\n')
l1=['May\n','June\n']
f1.writelines(l1)
f1.close()
```

```python
f2=open('f1.txt','r') #r= read mode,default mode, read from exiting file if not give an error
s=f2.read()
print(s)
s1=f2.read() # it will not print anything
print('o/p of s1 = ',s1)
f2.close()

f3=open('f2.txt')
s2=f3.read(3)
print(s2)
s3=f3.read(3) # \n also take as an charcter
print(s3)
f3.close()
```

```python
f4=open('f1.txt')
s4=f4.readlines() # print list
print(s4)
s5=f4.readlines() # empty list
print(s5)
f4.close()

f5=open('f2.txt')
s6=f5.readline() 
print(s6)
s7=f5.readline(2)
print(s7)
f5.close()

f6=open('f1.txt')
s8=f6.readlines(15)
print(s8)
f6.close()
```

```python
# wapp to count number of charter except \n
f=open('hello.txt')
c=0
for i in f.read():
    if i=='\n':
        continue
    elif i=='':
        break
    c+=1
print(c)
```
