read  data from text file
create dictionary in format
{movie title:{director: ,genre: ,budget: ,rating: }}
task-1 print average budget
task-2 find highest rated movies
task-3 count movies by genre
task-4 movies grouped by genre
task-5 drama movies sorted by rating(high to low)

```python
f=open("movies.txt")
m_d={}
avg_bud=0
c=0
for i in f:
    if i:
        parts=i.split(",")
        title=parts[0]
        director=parts[1]
        genre=parts[2]
        budget=float(parts[3])
        avg_bud+= budget
        rating=float(parts[4])
        m_d[title]={"director":director, "genre":genre, "budget":budget, "rating":rating}
        c+=1
        
print("movies dictionary: ",m_d)
# task2
print("average budget: ",avg_bud/c)

# task3
max=0
for i in m_d.values():
    if i['rating']>max:
        max=i['rating']
        
print("highest rated movie: ", max)

# task4
d_c={}
for i,j in m_d.items():
    genre=j["genre"]
    if genre not in d_c:
        d_c[genre]=1
    else:
        d_c[genre]+=1
for i,j in d_c.items():
    print(i,"-",j)

# task5
m_g={}
for i,j in m_d.items():
    genre=j["genre"]
    if genre not in m_g:
        m_g[genre]=[]
    m_g[genre].append(i)
for i,j in m_g.items():
    print(i,"-",j)
    
# task6
drama={}
for i,j in m_d.items():
    if j['genre']=="Drama":
        drama[i]=j['rating']
d_shorted=sorted(drama.items(),key=lambda x:x[1],reverse=True)
print(dict(d_shorted))
```
