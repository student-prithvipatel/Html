```python
import numpy as np
x=np.array([10,5])
y=np.array([30,20])
plt.plot(x,y,color="red",linestyle=":",linewidth=5)
plt.plot(x,y,"o")
plt.title("First Line Graph",fontsize=20,fontweight="bold",color="red")
plt.xlabel("x axis value",fontsize=10,fontweight="bold",color="green")
plt.ylabel("y axis value",fontsize=10,fontweight="bold",color="green")
plt.show()
```

```python
import numpy as np
x=np.array([10,5])
y=np.array([30,20])
z=np.array([20,10])
plt.plot(x,y,color="red",linestyle=":",linewidth=5)
plt.plot(x,y,"o",color="pink")
plt.plot(x,z,color="blue",linestyle="-.",linewidth=5)
plt.plot(x,z,"o",color="pink")
plt.title("First Line Graph",fontsize=20,fontweight="bold",color="red")
plt.xlabel("x axis value",fontsize=10,fontweight="bold",color="green")
plt.ylabel("y axis value",fontsize=10,fontweight="bold",color="green")
plt.show()
```

```python
# subplots
x=np.arange(1,11)
y1=2*x
y2=3*x
plt.subplot(1,2,1)
plt.title("Plot-1")
plt.plot(x,y1,"*:g")

plt.subplot(1,2,2)
plt.title("Plot-2")
plt.plot(x,y2,"^--r")
plt.suptitle("my graph")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
std=['a','b','c','d']
marks=[23,15,22,11]
plt.xlabel("Student")
plt.ylabel("marks")
plt.title("student vs marks")
plt.barh(std,marks,color=['red','blue','green','black'])
plt.show()
```

# Pie chart

```python
import matplotlib.pyplot as plt
x=[10,25,25,40,30]
l=['apple','banana','kiwi','cherry','mango']
plt.pie(x,labels=l,colors=['red','green','blue','yellow','pink'], 
       explode=[0.3,0,0,0,0],autopct="%0.1f%%")
plt.legend()
plt.show()
```

# area plot

```python
x=[1,2,3,4,5]
y=[1,4,8,6,2]
plt.plot(x,y,color="black")
plt.fill_between(x,y,alpha=0.3, color="red")
plt.show()
```

```python
x=[1,2,3,4,5]
y1=[1,4,8,6,2]
y2=[12,13,14,15,18]
y3=[22,35,45,32,23]
plt.stackplot(x,y1,y2,y3,labels=['a','b','c'],
             colors=['red','green','blue'],alpha=0.4)
plt.legend()
plt.show()
```

```python
month=np.arange(1,13)
income=np.array([45000,55000,48000,50000,51000,45000,55000,48000,50000,51000,56000,53000])
expense=np.array([51000,45000,55000,48000,50000,51000,56000,53000,45000,55000,48000,50000])
plt.plot(month,income,color="green")
plt.plot(month,expense,color="red")
plt.fill_between(month,income,expense,where=(income>expense),color="green",interpolate=True,label="positive")
plt.fill_between(month,income,expense,where=(income<=expense),color="red",interpolate=True,label="negative")
plt.legend()
plt.show()
```
