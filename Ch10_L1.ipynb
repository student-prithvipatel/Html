```python
import matplotlib.pyplot as plt
```

```python
x=[0,2,4,6,8]
y=[0,1,2,3,4]
plt.plot(x,y)
plt.show()
```

```python
import numpy as np
x=np.array([10,5])
y=np.array([30,20])
plt.plot(x,y,color="red",linestyle=":",linewidth=5)
plt.plot(x,y,"o")
plt.title("First Line Graph",fontsize=20,fontweight="bold",color="red")
plt.xlabel("x axis value",fontsize=10,fontweight="bold",color="green")
plt.ylabel("y axis value",fontsize=10,fontweight="bold",color="green")
plt.show()
```

```python
import numpy as np
x=np.array([10,5])
y=np.array([30,20])
z=np.array([20,10])
plt.plot(x,y,color="red",linestyle=":",linewidth=5)
plt.plot(x,y,"o",color="pink")
plt.plot(x,z,color="blue",linestyle="-.",linewidth=5)
plt.plot(x,z,"o",color="pink")
plt.title("First Line Graph",fontsize=20,fontweight="bold",color="red")
plt.xlabel("x axis value",fontsize=10,fontweight="bold",color="green")
plt.ylabel("y axis value",fontsize=10,fontweight="bold",color="green")
plt.show()
```

```python
y=np.array([0,1,2,3,4])
plt.plot(y,marker="*",ms=14,color="red")
plt.show()
```

```python
y=np.array([0,1,2,3,4])
plt.plot(y,"o:g") #r,b,c,m,w,k=black,y  linestlye- :,--,-.
plt.show()
```

```python
# subplots
x=np.arange(1,11)
y1=2*x
y2=3*x
plt.subplot(1,2,1)
plt.title("Plot-1")
plt.plot(x,y1,"*:g")

plt.subplot(1,2,2)
plt.title("Plot-2")
plt.plot(x,y2,"^--r")
plt.suptitle("my graph")

plt.show()
```

```python
x=np.arange(1,11)
y1=2*x
y2=3*x
plt.subplot(2,1,1)
plt.title("Plot-1")
plt.plot(x,y1,"*:g")

plt.subplot(2,1,2)
plt.title("Plot-2")
plt.plot(x,y2,"^--r")
plt.suptitle("my graph")

plt.show()
```
