468.  Write a function display_oddLines() to display odd number lines from the text file. Consider the following lines for the file – friends.txt.
 Friends are crazy, Friends are naughty !
 Friends are honest, Friends are  best !
 Friends are like keygen, friends are like license key !
 We are nothing without friends, Life is not possible without friends !

```python
f=open('friends.txt',)
s=f.readlines()
def display_oddLines():
    c=0
    for i in s: 
        c+=1
        if c%2==1:
            print(i)       
f.close()
        
display_oddLines()
```

472.  Write a python program to read line by line from a given files file1 & file2 and write into file3. 

```python
f1 = open("f1.txt",'r')
f2 = open("f2.txt",'r')
f3 = open("f3.txt",'w')
l1=len(f1.readlines())
l2=len(f2.readlines())
l=max(l1,l2)
f1.seek(0)
f2.seek(0)
for i in range(l):
    f3.write(f1.readline())
    f3.write(f2.readline())
f3.close()
f2.close()
f1.close()
```

474.  Write a python program to search for a string in text files.

```python
f = open("friends.txt",'r')
s=input("enter text you want to search from file: ")
l=f.readlines()
for i in l:
    if i.find(s)!=-1:
        print('line no= ',l.index(i)+1,'index no= ',i.index(s))
```
