```python
# task

import numpy as np
a1=np.array([[22,18,14,15],[23,15,22,18],[24,17,19,20],[16,19,13,21]])
etc=[22,23,24,24]
s5=[11,12,13,14,15]
s6=[10,9,6,5,14]

n=np.array(np.array_split(etc,4)) # n=etc.reshape(-1,1)
marks=np.concatenate((a1,n),axis=1)
print("marks with etc : \n",marks)

s5=np.array(s5).reshape(1,-1)
s6=np.array(s6).reshape(1,-1)
marks=np.concatenate((marks,s5,s6),axis=0)
print("marks after adding s5 and s6: \n",marks)

total= marks.sum(axis=1,keepdims=True)
print(total)

marks=np.concatenate((marks,total),axis=1)
print("marks after total: \n",marks)

print("after sorting:\n",np.array(sorted(marks,key=lambda x:x[-1], reverse=True)))
```
